# Add DBN columns
___

When we explored all of the data sets, we noticed that `class_size` does not have a `DBN` column. `hs_directory` has a `dbn` column that just needs to be renamed to an uppercase format.

```python
data['class_size'].head()
```

Comparing these with the rows from the other datasets such as `sat_results`, we can infer that the `DBN` is just the combination of the `CSD` and `SCHOOL CODE` columns. However, the `DBN` values are padded, which means that the `CSD` portion contains two digits. If the `CSD` is less than two digits, we will have to add a leading 0. 

To do this, we will create a custom function that 
1. Takes in a number
2. Converts it to a string
3. Checks its length
4. Adds a leading 0 if the length is 1.

The leading 0 is added using the `str.zfill()` method. 

The padded numbers will then be combined with the `SCHOOL CODE` values to create a `DBN` column for `class_size`.

```python
# Rename dbn column of `hs_directory`
data["hs_directory"].rename(columns={'dbn':'DBN'}, inplace=True) 

# Custom function that 'pads' the CSD numbers 
def pad_csd(num):
    string_num = str(num)
    if len(string_num) > 1:
        return string_num
    else:
        return string_num.zfill(2)
    
# Create a DBN column for class_size
data["class_size"]["padded_csd"] = data["class_size"]["CSD"].apply(pad_csd)
data["class_size"]["DBN"] = data["class_size"]["padded_csd"] + data["class_size"]["SCHOOL CODE"]
```

The first few rows of `class_size` are displayed to confirm the changes that we made. The new column should be at the end.

```python
data["class_size"].head()
```

# Combining the SAT Scores
___

As previously mentioned, we will want to sum up the three SAT scores into a single one to make our analysis easier. Working with a single number will make it easier to find correlations with the factors specified in our project. 

We will first convert the values of the three columns to numeric types with the `pandas.to_numeric()` method so that we can add them. The `errors=coerce` argument is placed so that pandas will treat unconvertible strings as missing values. 

```python
# Add the SAT scores to a single value after converting them to a numeric type
cols = ['SAT Math Avg. Score', 'SAT Critical Reading Avg. Score', 'SAT Writing Avg. Score']
for c in cols:
    data["sat_results"][c] = pd.to_numeric(data["sat_results"][c], errors="coerce")

data['sat_results']['sat_score'] = data['sat_results'][cols[0]] + data['sat_results'][cols[1]] + data['sat_results'][cols[2]]
```

# Parsing Geographic Coordinates of Schools
___

We noticed that `hs_directory` has a column, `Location 1`, that contains the geographic coordinates of latitude and longitude for each school. The first few cells of this column are displayed below. 

```python
pd.options.display.max_colwidth = 100 # To display all contents of each cell
data['hs_directory']['Location 1'].head()
```

The coordinates are located within parentheses at the end of each string. We will extract these values to map the schools which will then help us discover any geographic patterns within the data. 

To do this, we can use the vectorized string method `str.extract()` to extract each coordinate. Both will make use of the regular expression: `\(.+, .+\)` to isolate the coordinates portion of each cell in `Location 1`. Since parentheses have a special meaning in regular expressions, the backslashes cancel or escape this meaning and allows for the parentheses to be taken literally. The two `.+`s represent the latitude and longitude values which are separated by a comma, `,`. 

Capture groups (parentheses where each is NOT preceded by a backslash) will allow us to take either the latitude or the longitude, separately.  

```python
# Extract the latitudes and longitudes separately and store them in two new columns
data["hs_directory"]["lat"] = data["hs_directory"]["Location 1"].str.extract(r"\((.+), .+\)")[0] 
data["hs_directory"]["lon"] = data["hs_directory"]["Location 1"].str.extract(r"\(.+, (.+)\)")[0]

print(data["hs_directory"]["lat"].dtype)
print(data["hs_directory"]["lon"].dtype)
```

The `[0]` portion at the end of each line above is needed as the `str.extract()` method returns a DataFrame with a column named `0`. 

Since the `lat` and `lon` columns are of the type `object`, they are converted to numeric types below. 

The first few values for each are displayed to confirm our extraction process. 

```python
data["hs_directory"]["lat"] = pd.to_numeric(data["hs_directory"]["lat"], errors="coerce")
data["hs_directory"]["lon"] = pd.to_numeric(data["hs_directory"]["lon"], errors="coerce")

data["hs_directory"][['lat','lon']].head()
```

# Condense datasets
___

Since there are data sets with high schools that appear multiple times, this behavior carries over to the DBN values as well. Combining data sets without addressing this issue will result into problems. We would be merging a single row in one data set with two or more rows in another data set, which is not possible. 

Hence, we will have to condense the data sets with this multiplicity issue. Namely, these are:
- `class_size`
- `graduation`
- `demographics`

For `class_size`, the first few rows are displayed below.

```python
data['class_size'].head()
```

Unlike in `sat_results`, this data set contains non-high school information as indicated by the grade levels below. 

```python
data['class_size']['GRADE '].unique() # Take note of the space after GRADE
```

There are also different types of programs offered by each school. 

```python
data['class_size']['PROGRAM TYPE'].value_counts()
```

Since we are interested in high schools, we will select those rows with `09-12` for the `GRADE ` column. We will also choose `GEN ED` under `PROGRAM TYPE` as a second filter since the others are special types of educational programs. 

```python
# Filter class_size by GRADE and PROGRAM TYPE
class_size = data["class_size"]
class_size = class_size[class_size["GRADE "] == "09-12"]
class_size = class_size[class_size["PROGRAM TYPE"] == "GEN ED"]
class_size.head()
```

Above, we can see that there are still duplicate DBN values. This is due to the other columns such as `CORE SUBJECT (MS CORE and 9-12 ONLY)` and `CORE COURSE (MS CORE and 9-12 ONLY)`. It appears that these columns refer to different subjects and courses offered by each school. 

We want to have a single occurrence for each DBN value while somehow retaining the information contained in the data set above. Since we are more concerned of the numeric columns (that pertain to class size), we can condense the filtered data set by finding the averages of these columns. 

To do this, we will make use of the `pandas.DataFrame.groupby()` method. We will use this to group the data set by DBN value. We will then chain the `agg()` method, passing the `numpy.mean()`function as an argument (without parentheses), to compute the averages per DBN.  

```python
# Average the numeric columns to per DBN value
class_size = class_size.groupby("DBN").agg(np.mean)
class_size.reset_index(inplace=True)

data["class_size"] = class_size
data["class_size"].head()
```

We now have single occurrences of the DBN values for `class_size`. Note that the non-numeric columns are not included in the resulting dataset. 

Moving on to `demographics`, the first few rows are displayed below.

```python
data["demographics"].head(7)
```

We can see that there are multiple values for `schoolyear` per school. We will select the most recent school year, `20112012` which will also match our SAT results data set (2012). 

```python
data["demographics"] = data["demographics"][data["demographics"]["schoolyear"] == 20112012]
data["demographics"].head()
```

Finally, we move on to `graduation`. 

```python
data["graduation"].head()
```

We are concerned with the `Demographic` and `Cohort` columns, which refer to a specific demographic group and to the year of the data, respectively. 

```python
data["graduation"].Demographic.unique()
```

```python
data["graduation"].Cohort.unique()
```

Just like in `demographics`, we will want to select the latest year, which happens to be 2006. We will also select `Total Cohort` for the `Demographic` column to include all the students. 

```python
# Filter the graduation data set
data["graduation"] = data["graduation"][data["graduation"]["Cohort"] == "2006"]
data["graduation"] = data["graduation"][data["graduation"]["Demographic"] == "Total Cohort"]

data["graduation"].head()
```

We have now successfully condensed the data sets of `class_size`, `demographics`, and `graduation`.

# Combine the data sets
___

The `DBN` column will serve as the reference for our combining process. This will allow us to combine one row in a data set with another row in another data set if both contain the same DBN value. Do note, though, that DBN values may be present in one data set but absent in another. This is due to factors such as human error in data gathering, etc. 

We will use the `pandas.DataFrame.merge()` function to join our data sets. We need to be wary of which join type to use from `left`, `right`, `inner`, or `outer` as each one behaves differently and has its own pros and cons. Considering this behavior and the inconsistency of DBN values across the data sets, we cannot use the same join type for all data sets. Hence, we will combine the data sets gradually, using the appropriate join type as we go along. 

We start by creating a variable, `combined`, to contain our final data set. `sat_results` is placed there first as this is our primary data set. 

```python
combined = data["sat_results"]
```

We will perform the `left` joins first. This type of join allows us to retain all the rows of `sat_results`. This is most appropriate when combining our primary data set with those that have missing or fewer DBN values. Although this may cause us to lose rows from the other data set if there are DBN values there not found in `sat_results`, this is better than losing data from our primary data set. 

Below, we see that `ap_2010` and `graduation` fall under this criteria. 

```python
# Print the number of DBN values per data set
data_sets = ['sat_results', 'ap_2010', 'graduation']

for d in data_sets:
    print('{0}{1} {2}'.format(d,':', data[d].DBN.count()))
```

We now do the `left` merge process below. 

```python
combined = combined.merge(data["ap_2010"], on="DBN", how="left")
combined = combined.merge(data["graduation"], on="DBN", how="left")

combined.head()
```

```python
combined.shape
```

For the other data sets, we can see that there are more DBN values. These data sets also hold more valuable information for our analysis. 

```python
# Print the number of DBN values per data set
to_merge = ["class_size", "demographics", "survey", "hs_directory"]

for d in to_merge:
    print('{0}{1} {2}'.format(d,':', data[d].DBN.count()))
```

To retain as much information, we will use the `inner` join for these data sets. This join type allows us to retain rows that have DBN values common to the two data sets being combined at a given merging process. 

The final data set is displayed after to confirm the changes we made. 

```python
to_merge = ["class_size", "demographics", "survey", "hs_directory"]

for m in to_merge:
    combined = combined.merge(data[m], on="DBN", how="inner")
    
combined.head()
```

```python
combined.shape
```

We can notice that, after the `inner` join, the number of rows has decreased from the initial 479 to 363. This is because of the DBN values in `sat_results` that could not be found in the other data sets. This is fine for our purposes since we are interested in high-level correlations.

There are also several missing values in the data set. Some columns are entirely missing values. This is due to the previous cleaning and merging steps we did.

Pandas will not be able to find correlations if there are missing values. To handle this, we will replace `NaN` with the mean of the column, wherever possible. The rest will be replaced with 0.     

```python
# Replace missing values with the mean and with 0
combined = combined.fillna(combined.mean())
combined = combined.fillna(0)

combined.notnull().describe()
```

We can confirm from the output above that all missing values have now been replaced.
